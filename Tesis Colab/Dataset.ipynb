```python
df_local['Coronary Artery Disease (CAD)'] = df_local['Coronary Artery Disease (CAD)'].replace({'Ya': 1, 'Tidak': 0})
df_local['Hypothyroidism'] = df_local['Hypothyroidism'].replace({'Ya': 1, 'Tidak': 0})
df_local['Diabetes Mellitus (DM)'] = df_local['Diabetes Mellitus (DM)'].replace({'Ya': 1, 'Tidak': 0})
df_local['Hyperlipidemia'] = df_local['Hyperlipidemia'].replace({'Ya': 1, 'Tidak': 0})
df_local['Obesity Status'] = df_local['Obesity Status'].replace({'Ya': 1, 'Tidak': 0})
df_local['Hypertension'] = df_local['Hypertension'].replace({'Ya': 1, 'Tidak': 0})
df_local['Other Disease'] = df_local['Other Disease'].replace({'Ya': 1, 'Tidak': 0})
df_local['HBsAg'] = df_local['HBsAg'].replace({'Reaktif': 1, 'Non Reaktif': 0})
df_local['Anti HCV'] = df_local['Anti HCV'].replace({'Reaktif': 1, 'Non Reaktif':0})
df_local['Anti HIV'] = df_local['Anti HIV'].replace({'Reaktif': 1, 'Non Reaktif': 0})
df_local['Gender'] = df_local['Gender'].replace({'P': 1, 'L': 0})
display(df_local.head())
```

```python
df_local['Comorbidity'] = (df_local['Coronary Artery Disease (CAD)'] +
                           df_local['Hypothyroidism'] +
                           df_local['Diabetes Mellitus (DM)'] +
                           df_local['Hyperlipidemia'] +
                           df_local['Obesity Status'] +
                           df_local['Hypertension']+
                           df_local['Other Disease'])
df_local.head()
```

```python
df_local['Body Mass Index (BMI)'] = (df_local['Weight'] / ((df_local['Height'] / 100) * (df_local['Height'] / 100))).round(2)
display(df_local.head())
```

```python
import numpy as np

df_local['Total Body Water (TBW)'] = np.where(
    df_local['Gender'] == 0, # If Gender is 0
    2.447 - (0.09156 * df_local['Age']) + (0.1074 * df_local['Height']) + (0.3362 * df_local['Weight']),
    # If Gender is 1
    -2.097 + (0.1069 * df_local['Height']) + (0.2466 * df_local['Weight'])
)
```

```python
df_local['Lean Mass (LM) (%)'] = np.where(
    df_local['Gender'] == 0, # If Gender is 0 (Laki-laki)
    (((0.407 * df_local['Weight']) + (0.267 * df_local['Height'] ) - 19.2)/df_local['Weight'])*100,
    # If Gender is 1 (Perempuan)
    (((0.252 * df_local['Weight']) + (0.473 * df_local['Height']) - 48.3)/df_local['Weight'])*100
)

display(df_local.head())
```

```python

df_local['Total Body Fat Ratio (TBFR) (%)'] = (100 - df_local['Lean Mass (LM) (%)'])
display(df_local.head())
```

```python
# ========================================================================
# 1. Definisikan fungsi CKD-EPI eGFR
# ========================================================================
def eGFR_ckd_epi(Scr, age, gender):
    """
    Scr : serum kreatinin (mg/dL)
    age : usia pasien (tahun)
    gender : 0 = Male, 1 = Female
    """
    if gender == 1:  # Female
        kappa = 0.7
        alpha = -0.329
        gender_factor = 1.018
    else:  # Male
        kappa = 0.9
        alpha = -0.411
        gender_factor = 1

    min_ratio = min(Scr/kappa, 1)
    max_ratio = max(Scr/kappa, 1)

    egfr = 141 * (min_ratio**alpha) * (max_ratio**-1.209) * (0.993**age) * gender_factor
    return egfr

# ========================================================================
# 2. Terapkan fungsi ke DataFrame
# ========================================================================
# Pastikan df_local memiliki kolom 'Creatinine', 'Age', 'Gender' penghitungan maneggunakan eGFR
df_local['Glomerular Filtration Rate (GFR)'] = df_local.apply(lambda row: eGFR_ckd_epi(
    Scr=row['Creatinine'],
    age=row['Age'],
    gender=row['Gender']
), axis=1)

# ========================================================================
# 3. Lihat hasil
# ========================================================================
print(df_local[['Age','Gender','Creatinine','Glomerular Filtration Rate (GFR)']].head())
```

Feature Engineering

Rasio Lipid (Lemak Darah) Batu empedu kolesterol adalah jenis yang paling umum, dan pembentukannya terkait langsung dengan ketidakseimbangan lipid. Rasio ini lebih informatif daripada nilai absolut

```python
#TC/HDL Ratio
#Indikator risiko kardiovaskular dan dislipidemia umum. Rasio tinggi menunjukkan profil lemak yang buruk.
data['TC/HDL Ratio'] = data['Total Cholesterol (TC)'] / data['High Density Lipoprotein (HDL)']

#LDL/HDL Ratio
#Mengukur perbandingan "kolesterol jahat" terhadap "kolesterol baik." Rasio tinggi sangat erat kaitannya dengan risiko penyakit metabolik.
data['LDL/HDL Ratio'] = data['Low Density Lipoprotein (LDL)'] / data['High Density Lipoprotein (HDL)']

#Atherogenic Index (AI)
#Metrik alternatif untuk menilai risiko aterogenik, yang mencerminkan kolesterol non-HDL.
data['Atherogenic Index'] = (data['Total Cholesterol (TC)'] - data['High Density Lipoprotein (HDL)'])/data['High Density Lipoprotein (HDL)']

#Triglyceride/HDL Ratio (TG/HDL)
#Indikator yang kuat dari resistensi insulin dan dislipidemia aterogenik, faktor risiko penting dalam sindrom metabolik.
data['Triglyceride/HDL Ratio'] = data['Triglyceride'] / data['High Density Lipoprotein (HDL)']

data.head()
```

Indeks Morfologi Tubuh dan Komposisi Komposisi tubuh dan obesitas adalah faktor risiko utama batu empedu.

```python
#Non-Lean Mass (NLM)
#Perkiraan massa lemak tubuh total (meskipun TBFR sudah tersedia, ini memberikan perspektif yang berbeda menggunakan LM dan Weight).
data['Non-Lean Mass (NLM)'] = data['Weight'] - data['Lean Mass (LM) (%)']

#Body Fat/Water Ratio
#

data['Body Fat/Water Ratio']=data['Total Body Fat Ratio (TBFR) (%)']/data['Total Body Water (TBW)']
data.head()
```

**Indeks Fungsi Hati dan Ginjal** Batu empedu dipengaruhi oleh fungsi hati (tempat empedu diproduksi) dan ginjal (pengaruh pada ekskresi).

```python
#De Ritis Ratio (AST/ALT)
#Digunakan untuk menilai kerusakan hati. Ketidaknormalan hati dapat mempengaruhi komposisi empedu.
data['De Ritis Ratio'] = data['Aspartat Aminotransferaz (AST)'] / data['Alanin Aminotransferaz (ALT)']

data.head()
```
